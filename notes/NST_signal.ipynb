# MIT-BIH Noise Stress Test Database

NST database includes 12 half-hour ECG recordings obtained by using two clean recordings (118 and 119) from MIT-BIH Arrhythmia Database and adding noises typical in ambulatory ECG recordings such as baseline wander (in record 'bw'), muscle artifact (in record 'ma'), and electrode motion artifact (in record 'em').

You can downloard the database on [physionet](https://physionet.org/content/nstdb/1.0.0/).

## MIT-BIH

```python
%matplotlib inline

import wfdb
import numpy as np
import matplotlib.pyplot as plt

record = wfdb.rdrecord('../resources/mit-bih/118')
ann = wfdb.rdann('../resources/mit-bih/118','atr')

data = record.p_signal
t = 1/record.fs*np.arange(len(data))
ax1 = plt.subplot(2,1,1)
ax1.plot(t, data[:, 0])
ax1.set_ylabel(f'{record.sig_name[0]} ({record.units[0]})')
for i in range(len(ann.symbol)):
    if ann.symbol[i] in 'NLRBAaJSVrFejnE/fQ?':
        ax1.text(ann.sample[i]/record.fs, 0, ann.symbol[i], clip_on=True)
ax2 = plt.subplot(2,1,2, sharex=ax1)
ax2.plot(t, data[:, 1])
ax2.set_ylabel(f'{record.sig_name[1]} ({record.units[1]})')
ax2.set_xlabel('t (s)')
ax2.set_xlim([350, 360])
```
